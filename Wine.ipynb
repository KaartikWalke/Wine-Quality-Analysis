```python
#store wine type as an dim
red_wine['wine_type']="red"
white_wine['wine_type']="white"
#combine quality score to respective wines
red_wine['quality_label']=red_wine['quality'].apply(lambda value: 'low'if value<=5 
                                                    else 'medium' if value<=7
                                                    else 'high')
red_wine['quality_label']=pd.Categorical(red_wine['quality_label'],
                                        categories=['low','medium','high'])

white_wine['quality_label'] = white_wine['quality'].apply(lambda value: 'low' 
                                                              if value <= 5 else 'medium' 
                                                                  if value <= 7 else 'high')
white_wine['quality_label'] = pd.Categorical(white_wine['quality_label'], 
                                             categories=['low', 'medium', 'high'])
```

```python
corrMatrix=red_wine.corr()
```

```python
f,ax=plt.subplots(figsize=(9,8))
ax=sns.heatmap(corrMatrix,annot=True,cmap='icefire')
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2)
sns.countplot(x='quality_label',data=red_wine,ax=ax1)
sns.countplot(x='quality_label',data=white_wine,ax=ax2)
ax1.set_title("red wine")
ax2.set_title("white wine")
```

Here, we see bargraphs which show the number of samples of each quality of wine for both the types.
In the Dataset, we are having more number of low and medium quality wines than the high quality wines.
Also, we can see that the number of white wine samples is much more than the number of red wine samples.

MERGING THE DATASETS

```python
wines=pd.concat([red_wine,white_wine])
wines=wines.sample(frac=1,random_state=42).reset_index(drop=True)  
#frac=1 means that take 100% values if 0.5 means take 50%
wines.head()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2)
#fig.figsize=(18,18)
sns.boxplot(x="quality_label",y="alcohol",data=red_wine,ax=ax1)
sns.boxplot(x="quality_label",y="alcohol",data=white_wine,ax=ax2)
ax1.set_title("red wine")
ax2.set_title("white wine")
```

Here, we see that the boxplots are showing the alcohol concentration along with the quality label for all 3 types of qualities.

```python
f,(ax1,ax2)=plt.subplots(1,2,figsize=(10,4))
f.suptitle('wine quality -alcohol content/ph',fontsize=14)
f.subplots_adjust(top=0.80,wspace=0.3)

sns.boxplot(x="quality_label",y="alcohol",
           data=wines,ax=ax1)
ax1.set_xlabel("Red Wine Quality")
ax1.set_ylabel("wine Alcohol %")

sns.boxplot(x="quality_label",y="pH",
           data=wines,ax=ax2)
ax1.set_xlabel("Red Wine Quality",size=12)
ax1.set_ylabel("wine PH")
```

Here, we see that the pH levels in wines is shown along with quality label for all 3 types of qualities.
In Red Wine, there is greater pH in high quality wine than the low and medium quality wine.
In White Wine, there is greater pH in low quality wine than the high and medium quality wine.

```python
red_wine.hist(bins=20, color='red',edgecolor='black',
             linewidth=1.5,xlabelsize=8,
             ylabelsize=8,grid=False)
plt.tight_layout(rect=(0,0,1.2,1.2))
```

```python
white_wine.hist(bins=20, color='skyblue',edgecolor='black',
             linewidth=1.5,xlabelsize=8,
             ylabelsize=8,grid=False)
plt.tight_layout(rect=(0,0,1.2,1.2))
```

```python
g=sns.FacetGrid(wines,col="wine_type",hue="quality_label",
             col_order=['red','white'],hue_order=['low','medium','high'],
             aspect=1.2,size=3.5,palette=sns.light_palette('navy',3))
g.map(plt.scatter,"volatile acidity","alcohol",edgecolor="white",
     linewidth=0.5)
```

Here, the facetgrid shows the alcohol content along with increasing volatile acidity in both types of wine.
Dark Blue colour shows high quality and as the colour lightens to white, the dots signify medium and low quality.
We can see that high quaity wines have low volatile acidity on compared to medium and low quality wine.
We can also see that low quality wine have greater amount of volatile acidity as compared to higher quality wine. 

```python
sns.violinplot(x="quality",y="volatile acidity", hue="wine_type",
              data=wines, split=True,inner="quart",linewidth=1.5)
```

Here, in the Violinplot we see a comparison of volatile acidity along with quality between Red Wine and White Wine.
It shows that Red Wine has more volatile acidity than White Wine.

We can also see that volatile acidity decreases with increase in the quality of both Red Wine and White Wine.

```python
#joint plot
sns.jointplot(x='quality',y='sulphates',data=red_wine,
             kind='reg',ylim=(0,2),color='red',
             size=5.5,ratio=4)
```

Here, the jointplot we have drawn a representation of the amount of sulphates with the increase in quality of Red Wine. 

We can see that the amount of sulphates is more in medium quality than the low and high quality Red Wine.
